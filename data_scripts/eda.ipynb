```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
all_sites_df = pd.read_csv('../data/all_sites_df.csv', index_col=0)
```

```python
features = ['dswe', 'pr', 'temp', 'tmin', 'tmax', 'dah', 'trasp', 'elev', 'slopeGTOPO', 'aspectGTOPO']#, 'swe']
```

```python
features_df = all_sites_df[features]
```

# Correlation Matrix

```python
features_df_clean = features_df[(features_df['tmin'] < 3000) & (features_df['pr'] < 500) & (features_df['dswe'] < 400) & (features_df['dswe'] > -400)]
```

```python
plt.figure(figsize=(12, 10))
corr = features_df_clean.corr()
matrix = np.triu(np.ones_like(corr))
corr
sns.heatmap(corr, cmap='bwr', vmin=-1, vmax=1, annot=True, mask=matrix, xticklabels=features_names, yticklabels=features_names)
plt.tick_params(axis='x', labelsize=13)
plt.tick_params(axis='y', labelsize=13)
# plt.savefig(f'{save_dir}corr_heatmap.png', dpi=300)
```
